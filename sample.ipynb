```python
import pandas as pd
```

```python
sample = pd.read_json('dataset.json')
```

```python
df = pd.json_normalize(sample['features'])
```

```python
df_temp = df.apply(lambda col: col.map(str)) #fixing columns (containing lists or dicts )which cant be hashed for comparison by converting them to strings
```

```python
print(df.dtypes)
```

```python
print(df.columns.tolist())
```

```python
df['properties.WARD_NO'] = df['properties.WARD_NO'].astype(int)
df['properties.MOVEMENT_ID'] = df['properties.MOVEMENT_ID'].astype(int)
```

```python
print(df.dtypes)
```

```python
df.head()
```

```python
df.rename(columns={
    'properties.WARD_NO': 'ward_id',
    'properties.WARD_NAME': 'ward_name',
    'properties.MOVEMENT_ID': 'movement_id',
    'properties.DISPLAY_NAME': 'address'
}, inplace=True)

print(df.columns.tolist())
```

```python
df.head()
```

```python
print(df['geometry.type'].unique())
```

```python
df.drop(columns=['geometry.type'], inplace=True)
print(df.columns.tolist())
```

```python
df['ward_name'] = df['ward_name'].str.replace('.', '', regex=False)
df['address'] = df['address'].str.title()
df['address'] = df['address'].str.replace(r'^\d+\s', '', regex=True)
df['address'] = df['address'].str.replace('#', '', regex=False)
df['address'] = df['address'].str.replace("'", '', regex=False)
```
