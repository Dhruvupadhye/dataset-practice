```python
from shapely.geometry import shape
import json

with open('dataset.json') as f:
    data = json.load(f)
```

```python
df['area'] = [shape(f['geometry']).area for f in data['features']]
```

```python
df['mapped_regions'] = df['address'].str.split(',').str.len()
```

```python
df.head()
```

```python
df['latitude'] = [shape(f['geometry']).centroid.y for f in data['features']]
df['longitude'] = [shape(f['geometry']).centroid.x for f in data['features']]
```

```python
df.head()
```

```python
df.drop(columns=['type', 'movement_id', 'geometry.coordinates'], inplace=True)
```

```python
import folium
import geopandas as gpd
import json

gdf = gpd.read_file('dataset.json')


gdf['area'] = df['area'].values
gdf['mapped_regions'] = df['mapped_regions'].values
gdf['latitude'] = df['latitude'].values
gdf['longitude'] = df['longitude'].values


m = folium.Map(location=[12.97, 77.59], zoom_start=11)


def get_color(area):
    if area > 0.001:
        return 'red'
    elif area > 0.0005:
        return 'orange'
    elif area > 0.0002:
        return 'yellow'
    else:
        return 'green'


for _, row in gdf.iterrows():
    folium.GeoJson(
        row['geometry'].__geo_interface__,  
        style_function=lambda x, area=row['area']: {
            'fillColor': get_color(area),
            'color': 'black',
            'weight': 1,
            'fillOpacity': 0.6
        },
        tooltip=f"Ward: {row['WARD_NAME']} | ID: {row['WARD_NO']} | Area: {row['area']:.6f} | Mapped Regions: {row['mapped_regions']}"  # ← fix: simple string tooltip
    ).add_to(m)


top10_area = df.nlargest(10, 'area')
for _, row in top10_area.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=folium.Popup(
            f"<b>{row['ward_name']}</b><br>Area: {row['area']:.6f}<br>Mapped Regions: {row['mapped_regions']}",
            max_width=200
        ),
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(m)


legend_html = """
<div style="position: fixed; bottom: 30px; left: 30px; z-index: 1000;
     background-color: white; padding: 10px; border-radius: 5px;
     border: 2px solid grey; font-size: 14px;">
     <b>Ward Area Size</b><br>
     <i style="background:red;width:12px;height:12px;display:inline-block"></i> Largest (>0.001)<br>
     <i style="background:orange;width:12px;height:12px;display:inline-block"></i> Medium-Large (>0.0005)<br>
     <i style="background:yellow;width:12px;height:12px;display:inline-block"></i> Medium (>0.0002)<br>
     <i style="background:green;width:12px;height:12px;display:inline-block"></i> Smallest (<0.0002)<br>
</div>
"""
m.get_root().html.add_child(folium.Element(legend_html))


m.save('bengaluru_wards_map.html')
print("Map saved!")
```
